```python
import pandas as pd
import numpy as np

print("Libraries loaded successfully")
```

```python
# Data from FSI 2021
# columns : district, geo area , forest cover , type
data = {
    'district': [
        'Balaghat', 'Mandla', 'Dindori', 'Seoni', 'Chhindwara',
        'Betul', 'Hoshangabad', 'Shivpuri', 'Morena', 'Bhind',
        'Jabalpur', 'Katni', 'Umaria', 'Shahdol', 'Anuppur',
        'Rewa', 'Satna', 'Damoh', 'Sagar', 'Panna',
        'Indore', 'Bhopal', 'Gwalior', 'Ujjain', 'Ratlam'
    ],
    'geo_area': [
        9245, 8771, 7470, 8758, 11815,
        10043, 6707, 10278, 4989, 4459,
        5211, 4947, 4548, 6205, 3701,
        6314, 7502, 7306, 10252, 7135,
        3898, 2772, 4560, 6091, 4861
        ],
    'forest_cover': [
        4437, 4163, 3936, 3619, 3474,
        2981, 1672, 1421, 312, 88,
        1510, 1824, 2489, 2741, 2103,
        980, 1562, 1870, 1943, 2673,
        190, 298, 141, 280, 456
    ],
    'region': [
        'Central', 'Central', 'Central', 'Central', 'Central',
        'Central', 'Central', 'North', 'North', 'North',
        'Central', 'Central', 'East', 'East', 'East',
        'East', 'East', 'Central', 'Central', 'Central',
        'West', 'Central', 'North', 'West', 'West'
    ]
}
df = pd.DataFrame(data)
df['forest_pct'] = (df['forest_cover'] / df['geo_area'] * 100).round(1)
print(df.shape)   # rows, columns
df.head()
```

```python
df[['forest_cover', 'forest_pct']].describe().round(1)
# describe() is UNDER NUMPY helps to understanfd summary
```

```python
top5 = df.sort_values('forest_pct', ascending=False).head(5)
bottom5 = df.sort_values('forest_pct').head(5)

print("Top 5 forest districts:")
print(top5[['district', 'forest_pct']])
print("Bottom 5 forest districts:")
print(bottom5[['district', 'forest_pct']])

# sort values function is used and it comes under pandas called pandas sorting
```

```python
region_stats = df.groupby('region').agg(
    total_forest = ('forest_cover', 'sum'),
    avg_pct      = ('forest_pct', 'mean'),
    districts    = ('district', 'count')
).round(1)
print(region_stats)
# group by is a pandas function
```

```python
pct_array = df['forest_pct'].values  # convert to numpy array

print(f"Mean   : {np.mean(pct_array):.1f}%")
print(f"Median : {np.median(pct_array):.1f}%")
print(f"Std Dev: {np.std(pct_array):.1f}%")
print(f"Max    : {np.max(pct_array):.1f}%")
print(f"Min    : {np.min(pct_array):.1f}%")

# Boolean filtering with numpy
above_33 = np.sum(pct_array > 33)
print(f"\nDistricts above 33% target: {above_33} out of {len(pct_array)}")
#here we find the mean , median, max, min by converting to numpy array
```

```python
df['category'] = np.where(
    df['forest_pct'] >= 33, 'Good',
    np.where(df['forest_pct'] >= 20, 'Moderate', 'Low')
)

# value_counts() — count each category
print(df['category'].value_counts())
# np.where islike and if-else for arrays and we see the type also
```

```python
df.to_csv('mp_forest_analysis.csv', index=False)
print("File saved! Shape:", df.shape)
print("Columns:", df.columns.tolist())
# heres the analysis
```
